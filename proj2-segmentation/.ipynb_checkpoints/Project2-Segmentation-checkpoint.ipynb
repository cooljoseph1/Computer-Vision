## Graph Cut:
Your code should read in an image and a set of seed pixels and use graph-cut segmentation to segment the image.

You will need to calculate the costs for the t-links (region terms) and the n-links (boundary terms). See the book, the notes/slides, or published papers in this area for ideas of how to define these.  Remember that the t-link weights to a particular terminal (foreground or background) should be large if that pixel looks a lot like the respective foreground/background seeds. The n-link weights should be large if the two neighboring pixels are similar.

Here is [the original paper on graph-cut segmentation](http://www.csd.uwo.ca/~yuri/Papers/iccv01.pdf), which might help with some ideas, but you should look at the literature to see what other costs functions / link weights others have used.

Once the graph is built, use the min-cut algorithm to partition the graph into nodes connected to the foreground node or to the background node, then use these as the resulting labels for the segmentation. Display this result graphically in some fashion overlaid on the input image.  It is best to start with simple images whose foreground and background colors are pretty different and for ones where the edges are pretty clear.  Graph-cut segmentation struggles sometimes with long, thin structures, so you should avoid these types of images early on.



````python
%pylab inline
import guiseg
from PIL import Image
from sklearn.neighbors import KernelDensity
import numba
import maxflow
````

````python
def manhattan_distances(image, sources):
    """
    Generate an array of manhattan distances from the source array
    """
    height, width = image.shape[:2]
    result = zeros((height + 1, width + 1)) + 500
    result[sources] = 0
    @numba.njit()
    def update_result(result):
        # Pass forward:
        for y in range(1, height):
            for x in range(1, width):
                result[y, x] = min(result[y, x], min(result[y-1, x], result[y, x-1]) + 1)

        # Pass backwards:
        for y in range(height-1, -1, -1):
            for x in range(width-1, -1, -1):
                result[y, x] = min(result[y, x], min(result[y+1, x], result[y, x+1]) + 1)
        return result
    result = update_result(result)
    
    return result[:-1, :-1]
````

````python
def create_kde(pixels, bandwidth=10):
    kde = KernelDensity(bandwidth=bandwidth)
    kde.fit(pixels)
    return kde
````

````python
def create_graph(image, source_kde, sink_kde, source_distances, sink_distances):
    """
    Create the graph
    """
    image = image.astype(float)
    height, width = image.shape[:2]
    # Create graph structure:
    graph = maxflow.Graph[float](height * width, 2 * height * width - height - width)
    nodes = graph.add_nodes(height * width)
    
    # Create n-links (boundary term)
    n_lambda = 200
    n_sigma = 30
    n_link_vertical = n_lambda * np.exp(-np.sum(np.square((image[1:] - image[:1])), axis=2) / 2 / n_sigma**2)
    for y in range(height - 1):
        for x in range(width):
            graph.add_edge(nodes[y*width + x],
                           nodes[y*width + x + width], 
                           n_link_vertical[y, x],
                           n_link_vertical[y, x])
            
    n_link_horizontal = n_lambda * np.exp(-np.sum(np.square((image[:, 1:] - image[:, :1])), axis=2) / 2 / n_sigma**2)
    for y in range(height):
        for x in range(width - 1):
            graph.add_edge(nodes[y*width + x], 
                           nodes[y*width + x + 1], 
                           n_link_horizontal[y, x], 
                           n_link_horizontal[y, x])
    
    # Create t-links (region term)
    distances_lambda = 0.05
    t_link_source = distances_lambda * source_distances.reshape(-1) - source_kde.score_samples(image.reshape(-1, 3))
    t_link_sink = distances_lambda * sink_distances.reshape(-1) - sink_kde.score_samples(image.reshape(-1, 3))

    for i in range(len(t_link_source)):
        graph.add_tedge(nodes[i], t_link_source[i], t_link_sink[i])

    graph.maxflow()
    return graph, nodes
````

````python
def get_segment(image, graph, nodes, segment=0):
    height, width = image.shape[:2]
    segmented_image = zeros((height, width))
    for y in range(height):
        for x in range(width):
            segmented_image[y, x] = segment ^ graph.get_segment(nodes[y * width + x])
    return segmented_image
````

## Write up

My algorithm worked as follows:
1. Get user input for foreground and background using the guiseg.py module.
2. Create a kernel density estimator and fit it to the foreground/background pixels.  The kernel density estimator is implemented using sklearn.neighbors.KernelDensity and basically adds a Gaussian to each provided input to create what looks similar to a three dimensional histogram, except smooth.
3. Compute the Manhattan distances for each point on the image to the closest foreground/background point.
4. Create a graph and use pymaxflow to find the min cut.  The min cut segments the image into two pieces -- the desired foreground and background.
    
To create the graph, I made each pixel in the image a node.  Each pixel was connected to its neighboring pixels and a source and sink node.  The connections between neighboring pixels are called n-links, and the connections to the source or sink are called t-links.

The formula I used for n-links was
$$\text{n-link}(\mathbf{x_1}, \mathbf{x_2}) := \lambda_n e^{-\frac{\|\mathbf{x_1} - \mathbf{x_2}\|^2}{2\sigma^2}}$$
where $\mathbf{x_1}$ and $\mathbf{x_2}$ are the color vectors for the corresponding pixels, and $\lambda_n$ and $\sigma$ are provided variables (set to $200$ and $30$, respectively).  My computations for n-links were vectorized to improve speed.

To compute t-links, I took the Manhattan distance from the source (or sink), multiplied this by some constant, $\lambda_t$, and subtracted off the natural log of the probability that it belongs to the source (or sink) based on the kernel density estimator.  In pseudocode, this is:
$$\text{t-link}(\text{pixel}) := \lambda_t \cdot \text{closest_manhattan_distance}(\text{pixel}, \text{foreground pixels}) - \ln(\text{foreground_kde.score_sample(pixel)})$$
The point of the Manhattan distances is to penalize pixels a little from being far from a given foreground (or background) pixel.  The Manhattan distances can be rapidly computed by two passthroughs of the image pixels, which is why I chose that metric.  The point of the kernel density estimator is to penalize a pixel for being a wildly different color than the given foreground (or background) pixels.

My algorithm worked best on images that had high contrast between the background and foreground and for which the object was convex.  It had difficulty on blurry images, images with shadows, or images with many similar colors in the background and foreground.  For example, the algorithm works very well on the apple -- a simple, convex shape with high contrast -- and the teddy bear, which also has high contrast.  However, it did not work very well on the penguin or the log.  For the penguin, the colors in the background and foreground were quite similar;  the white on the penguin's stomach matched the white of the snow, and the black on its flippers, tail, and head matched the black of the rock.  The log is difficult even for a human to precisely segment.  Its shadow blurs it into the pond, and it is a similar color to much of the surrounding image, expecially the top left.

As you can probably guess, I felt like the apple I provided was an easy image, and the log was a difficult one.  The apple was easy because it had fairly uniform colors throughout the foreground and background, and had high contrast between the two.  The log was difficult because it did not have very good contrast between background and foreground, and many of the same colors were used in both the foreground and the background.

I learned a lot from this project, including how to rapidly compute Manhattan distances, what a kernel density estimator is and how to implement it, and the max-flow min-cut theorem.
