# Simple Hough transform example

```python
%pylab inline
from PIL import Image
from skimage.filters import sobel_h, sobel_v, gaussian
from skimage.color import hsv2rgb
rcParams['figure.figsize'] = (8,8)
```

We're going to perform the hough transform on this simple circle image:

```python
img = Image.open('simplecircle.png')
img
```

First, we need to find appropriate edges points to work with. I'm use a Sobel filter to calculate the gradient magnitude, then threshold the gradient magnitude to get edge points. I played around with the threshold until I found something that seemed reasonable.

```python
arr = array(img.convert('L'))  # numpy array version of the image
gy = sobel_h(arr)  # y gradient
gx = sobel_v(arr)  # x gradient
grd = sqrt(gx**2 + gy**2)  # gradient magnitude
grd /= grd.max()  # normalize the gradient magnitude to be in the range [0,1]
edges = grd > .1  # threshold the gradient magnitude to get edges
ori = arctan2(gy, gx) / (2*pi) + .5  # calculate the gradient orientation
orihsv = where(edges, ori, 0)  # only look at the orientation at the edge points
orihsv = hsv2rgb(stack([orihsv, ones_like(orihsv), where(edges, 1, 0)], 2))  # make the image pretty
fig, ax = subplots(1, 2, figsize=(10, 5), constrained_layout=True)
ax[0].imshow(edges)
ax[0].grid()
ax[1].imshow(orihsv)
```

Next, we need to define a circle "stamp" that we can use for voting in the accumulator array. I do this by making a large circle with a thick border, and then downsampling the image to the correct size. This yields a circle with a nice, antialiased border.

```python
r = 32  # radius of the circle
f = 24  # upscaling factor for making the "big" circle
r1 = r * f + 3  # outer radius of the thick border
r2 = r * f - 3  # inner radius of the thick border
y, x = mgrid[-r1-8:r1+9, -r1-8:r1+9]  # create a grid of x and y coordinates
big_circ = logical_and((x**2+y**2) <= r1**2, (x**2+y**2) >= r2**2)  # create the thick border big circle
sm_circ = asarray(Image.fromarray(big_circ.astype(uint8)*255).resize((2*r+1,2*r+1), Image.BICUBIC))  # resize to a small circle, using PIL Image.resize
sm_circ = sm_circ / sm_circ.max()  # renormalize the values of the circle to the range [0,1]
imshow(sm_circ)
```

Now we will perform the voting. We add a circle into the accumulator centered at the position indicated by each edge point. Note that we need to pad the accumulator to account for centers of circles being outside the bounds of the image. There are a couple different ways we could add the circle to the image. One is the "stamping" method. We simply add the stamp array to a slice of the accumulator array. We get that slice by doing something along the lines of `accumulator[y-r:y+r+1, x-r:x+r+1]`, where `x` and `y` are the position of the edge point, appropriately shifted for the accumulator. This is done for every edge point (probably in a loop).

Another method is the one I'm doing here. It's a little more challenging conceptually. Instead of stamping the entire circle array (which includes a bunch of zero values that I don't care about), I instead calculate the offset of each non-zero stamp value from the center of the stamp image. If I then add those offsets to the coordinates of any particular edge point, I get the indices that define a circle around that edge point. I can use those indices to add the values of the circle into the accumulator. Furthermore, because of numpy broadcasting and the use of `np.add.at`, I can do this for all edge points in the image at once, without having to write a loop. If you find it hard to understand what's going on here, you can try playing with the code, printing out intermediate variables, etc.

```python
epy, epx = edges.nonzero()  # x and y coordinates of all nonzero values in edges
cy, cx = sm_circ.nonzero()  # x and y coordinates of all nonzero values in the circle stamp
v = sm_circ[cy, cx]  # the actual nonzero values from the circle stamp
# make cx and cy relative to the center of the circle
cx -= r
cy -= r

# NOTE: cx, cy, and v all have the same shape. This is how they fit together:
# sm_circ[y[i], x[i]] = v[i]

# create the padded accumulator array
h, w = arr.shape
accum = zeros((h+2*r, w+2*r))

# for every edge coordinate, add the offsets of the circle. So xx, for instance, will be an array
# with shape (num_edge_points, num_offset_points), and each row will contain the x values for each
# point in the accumulator that make a circle centered at that edge value.
xx = epx[:, None] + r + cx
yy = epy[:, None] + r + cy
np.add.at(accum, (yy, xx), v)  # accumulate the values in v into accum at the coordinates given by yy and xx
imshow(accum)
```

The final step is to locate the max values in the accumulator. In this simple example, there is only one circle and no noise, so we should just be able to find the accumulator location with the maximum value. In more complex images, we have to do some careful processing of the accumulator array to identify likely circles.

```python
# get the coordinates of the max value in the accumulator
yp, xp = np.unravel_index(accum.argmax(), accum.shape)

# plot the location of the max value, and a circle centered at it
fig, ax = subplots(1,1)
ax.imshow(pad(img, ((r,r),(r,r),(0,0))), alpha=.5)
ax.scatter(xp, yp, color='w', marker='*')
gca().add_artist(Circle((xp, yp), r, ec='w', fc=(0,0,0,0)))
```
